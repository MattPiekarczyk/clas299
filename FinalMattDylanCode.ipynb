```scala
import scala.io.Source
val beazleyGroup = "https://raw.githubusercontent.com/neelsmith/clas299/master/vases-2020.cex"
val  pioneerGroup = "https://raw.githubusercontent.com/MattPiekarczyk/clas299/master/pioneerGroupPainters.cex"
```

First, we create a histogram of the shapes of both the Beazley painters as well as the pioneer group. We will explore the distribution of the most popular shape of vase, which we will find after running the code.

```scala
val lines = Source.fromURL(beazleyGroup).getLines.toVector
val data = lines.tail.map(ln => ln.split("#").toVector)
val shapeStrings = data.map(columns => columns(3))
val groupShapes = shapeStrings.groupBy(v => v)
val groupedShapesVector = groupShapes.toVector
val counts = groupedShapesVector.map{ case (site,examples) => (site, examples.size) }
counts.mkString("\n")
val shapesHisto =  counts.toVector.sortBy(frequency => frequency._2).reverse
```

So, we see hydriai was the most popular vase shape. Now, we do the same for the Pioneer Group:

```scala
val linesPioneer = Source.fromURL(pioneerGroup).getLines.toVector
val dataPioneer = linesPioneer.tail.map(ln => ln.split("#").toVector)
val shapeStringsPioneer = dataPioneer.map(columns => columns(3))
val groupShapesPioneer = shapeStringsPioneer.groupBy(v => v)
val groupedShapesVectorPioneer = groupShapesPioneer.toVector
val countsPioneer = groupedShapesVectorPioneer.map{ case (site,examples) => (site, examples.size) }
counts.mkString("\n")
val shapesHistoPioneer = countsPioneer.toVector.sortBy(frequency => frequency._2).reverse
```

We see, again, that hydriai is again the most popular vase shape. Now, we filter our data to only view these shapes. 

```scala
val filteredBeazley = data.filter(columns => columns(3) == "Hydriai")
val filteredPioneer = dataPioneer.filter(columns => columns(3) == "Hydriai")
```

Now that all of our data includes only Hydriai, we can see the frequency of each find spot.

```scala
val locationStringsPioneer = filteredPioneer.map(columns => columns(4))
val groupLocationsPioneer = locationStringsPioneer.groupBy(v => v)
val groupedLocationsVectorPioneer = groupLocationsPioneer.toVector
val countsLocationsPioneer = groupedLocationsVectorPioneer.map{ case (site,examples) => (site, examples.size) }
counts.mkString("\n")
val locationsHistoPioneer = countsLocationsPioneer.toVector.sortBy(frequency => frequency._2).reverse
```

```scala
val locationStringsBeazley = filteredBeazley.map(columns => columns(4))
val groupLocationsBeazley = locationStringsBeazley.groupBy(v => v)
val groupedLocationsVectorBeazley = groupLocationsBeazley.toVector
val countsLocationsBeazley = groupedLocationsVectorBeazley.map{ case (site,examples) => (site, examples.size) }
counts.mkString("\n")
val locationsHistoBeazley = countsLocationsBeazley.toVector.sortBy(frequency => frequency._2).reverse
```
